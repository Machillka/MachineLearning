## 可视化卷积操作

学习 tensorflow 提供的卷积层 / 池化层 API

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

%matplotlib inline
```

```python
# read the image

def ShowImg(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)

img = cv.imread('../Database/nagisa.png')

ShowImg(img)
```

```python
shaperKernel = np.array([
    [-1, -1, -1],
    [-1, 9, -1],
    [-1, -1, -1],
], dtype = np.float32)

verticalKernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1],
], dtype = np.float32)

# 将 verticalKernel 进行转置之后就成为了对应需要检测方向的滤波器
horizontalKernel = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1],
], dtype = np.float32)


yourKernel = np.array([
    [1.3, 1.34, 3.74],
    [1.3, 2.78, 3.74],
    [1.55, 1.34, 3.74],
], dtype = np.float32)

kernels = []

kernels.append(shaperKernel)
kernels.append(verticalKernel)
kernels.append(horizontalKernel)
kernels.append(yourKernel)

operatorImg = cv.filter2D(img, -1, yourKernel)

ShowImg(operatorImg)
```
